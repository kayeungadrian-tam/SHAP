## 概要
タイタニックでは、約900名分の乗客のデータが与えられますが、それぞれの乗客の属性（年齢、性別など）が入っています。一部欠損値がありますので、それらを補完しながら、データが与えられていない乗客に対して、生存しているか、生存していないかを判別するモデルを作るというものです。

```python
# デートの読み込み
train_df = pd.read_csv("data/train.csv")
```

```python
# 生値
display(train_df.head())
```

```python
# データ型
display(train_df.info())
```

```python
# 欠損値
display(pd.isna(train_df).sum())
```

```python
# ユニーク数
display(train_df.nunique())
```

## データ型の変換
機械学習を行うため、文字列のものを数値へ変換する。

```python
# SexとEmbarkedのユニーク値を確認する
for col in ["Sex", "Embarked"]:
    print(f"{col}:\t{train_df[col].unique()}")
```

```python
# それぞれの値を数値へ変更
train_df["Embarked"] = train_df["Embarked"].map({"Q":0, "S":1, "C":2})
train_df["Sex"] = train_df["Sex"].map({"male":0, "female":1})
display(train_df.info())
```

## 欠損値の処理
欠損データをダミーデータに入れ替える。

```python
# ここでは"median"の値に入れ替えるが、実際は様々な手法があり、
# ケース毎に判断する必要がある
train_df["Age"] = train_df["Age"].fillna(train_df["Age"].median())
train_df["Embarked"] = train_df["Embarked"].fillna(train_df["Embarked"].median())
train_df.isnull().any()
```

## 説明・目的変数の定義
Dataframeから説明変数（X）と目的変数（Y）を抽出する。

### 仮説

1. 性別が一番影響するのでは？
1. チケットクラス（Pclass）が次に影響するので？

```python
# 目的変数と必要なさそうなデータを除外
X = train_df.drop(["Survived","Cabin", "Ticket", "Name", "PassengerId"], axis=1)
X.info()
```

```python
# 目的変数を抽出
y = train_df["Survived"]
y.shape
```

```python
# Shapeの確認
display(y.shape, X.shape)
```

## Train & Test
訓練用データセットとテスト用データセットの２つに分ける。

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=329)
print("X_train: ", X_train.shape)
print("y_train: ", y_train.shape)
print("X_test:  ", X_test.shape)
print("y_test:  ", y_test.shape)
```

## モデル生成
今回のデータセットはテーブルデータにおける２値分類タスクであるため、決定木をBaselineとする。

```python
# モデル生成およびパラメータ確認
RFC = RandomForestClassifier()
display(RFC.get_params())
```

```python
# 学習を行う
RFC.fit(X_train, y_train)
```

```python
# テストデータに対して予測を行う
y_pred = RFC.predict(X_test)
y_pred_prob = RFC.predict_proba(X_test)[:,1]
```

```python
# 簡易的な評価
metrics.accuracy_score(y_test, y_pred)
```

# SHAPについて
 シャープレイ値（SHAP）とは、ゲーム理論に基づいた、XAIに属する手法である。


```python
# SHAPの定義～SHAP値の算出
exp=shap.TreeExplainer(RFC)
sv_test = exp.shap_values(X_test)
# sv_testは[負と予測する方向への寄与の平均、正と予測方向への寄与の平均]
sv_test = sv_test[1]
sv_test.shape
```

```python
# 0行目の特徴量の寄与
sv_test[0]
```

```python
# SHAPの一致性
exp.expected_value[1] + sv_test[0].sum(), y_pred_prob[0]
```

```python
sv_sum = sv_test.sum(axis=1) + exp.expected_value[1]
plt.scatter(sv_sum, y_pred_prob, s=1)
plt.xlabel("Shap sum")
plt.ylabel("Pred Porb")
plt.savefig("fig/Pred_prob.png")
```

# SHAPを用いたXAI
Shap値を用いてモデルを説明する。

```python
# 単体の説明
i = 21
shap.force_plot(exp.expected_value[1], sv_test[0], X_test.iloc[i], matplotlib=True)
```

```python
# 全ての特徴量の影響
shap.decision_plot(exp.expected_value[1], sv_test[i], X_test.iloc[i])
```

```python
# 全体の特徴量
fig = shap.summary_plot(sv_test, X_test, show=False)
display(fig)
```

```python
# RFモデルにおける全ての特徴量の影響の寄与
fig = shap.summary_plot(sv_test, X_test, plot_type="bar", show=False)
plt.savefig("fig/summary_bar.png")
```

```python
# 特徴量毎の影響
feat_list = X_test.columns

for idx, v in enumerate(feat_list):
    fig = shap.dependence_plot(v, sv_test, X_test, show=False)
    plt.savefig(f"fig/depend_{v}.png")
```
